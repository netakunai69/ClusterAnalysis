```python
import pandas as pd
from scipy.cluster.vq import kmeans2, whiten
%matplotlib inline

dummy = pd.read_excel('kmeans.xlsx')

#kmeansに適用できうようにデータを正規化
whitened = whiten(dummy)
#K=2でkmeans法を適用
result = kmeans2(whitened, 2)

#kmeans法の結果を確認
import matplotlib.pyplot as plt
plt.plot(whitened[result[1]==0,0],whitened[result[1]==0,1],'ob',
     whitened[result[1]==1,0],whitened[result[1]==1,1],'or')
plt.plot(result[0][:,0], result[0][:,1],'sm',markersize=8)
```
