```python
#Excelデータを解析で扱えるようにする
#データ読み込みのためのpandasライブラリをロードする
import pandas as pd
#Excelファイルをcosmeという名前で読み込む
cosme = pd.read_excel('cosmetics.xlsx')
#cosmeの中身を表示する
cosme
```

```python
#必要なライブラリのロード
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
#Jupyterでプロットするための宣言
%matplotlib inline
#データMを最短距離法でクラスタリング
result = linkage(cosme, method='single')
#デンドログラムの作成
dendrogram(result, color_threshold=0)
plt.show()
```

```python
result
```

```python
result = linkage(cosme, method="ward")
dendrogram(result, color_threshold=0)
plt.show()
```

```python
result
```

```python
from scipy.cluster.hierarchy import cut_tree
cut_tree(result, height=12)
```
